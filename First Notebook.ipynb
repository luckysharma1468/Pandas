```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df
```

```python
# how argument values 
# 1. any -> if 'any' values is NAN ---> Drop that row and column (By Default it is True)
# 2. all -> if 'all' values is NAN ---> then only remove the row and column
 
df.dropna()
```

```python
df
```

```python
# Modifying the DF dataframe

df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, np.nan, np.nan],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df
```

```python
# If axis given then that axis is considered only and according to any and all

# any is used with axis = 1, any value in that column is null then whole column will be removed

df.dropna(how='any', axis=1)
```

```python
# all is used with axis = 1, if all the values is NAN then only column will be removed

df.dropna(how='all', axis=1)
```

```python
# Modifying the DF dataframe

df = pd.DataFrame({"name": [np.nan, 'Batman', 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df
```

```python
# No axis is provided then row and column both is considered

df.dropna(how="all")
```

```python
df
```

```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Alfred'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df
```

```python
# Dropping duplicates

# taking subset as a argument ---> name of the column passed as a list

# keep argumnet ---> 3 values
# 1. keep = first --> keep the first occurence (By default = True)
# 2. keep = last ---> keep the last occurrence
# 3. keep = False ---> remove all the occurences

df.drop_duplicates(subset = ['name'])
```

```python
df
```

```python
# keeping the last occurrence of the duplicate value

df.drop_duplicates(subset = ["name"], keep = "last")
```

```python
# Removing all the duplicates

df.drop_duplicates(subset = ["name"], keep = False)
```

```python
# All the drop duplicates operations are on a copy so orignal data frame is intact

df
```
